```python
df.dropna(how='any', inplace=True)
df.info()
```

```python
from sklearn import preprocessing

def encode_feature(df, col_list):
    for feature in col_list:
        le = preprocessing.LabelEncoder()
        le = le.fit(df[feature])
        df[feature] = le.transform(df[feature])

    return df
```

```python
encode_feature(df, ['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'Property_Area', 'Loan_Status'])
```

```python
df
```

```python
df.drop('Loan_ID', axis=1, inplace=True)
```

```python
df.info()
```

```python
y = df['Loan_Status']
df.drop('Loan_Status', axis=1, inplace=True)

df.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.2, random_state=42)
```

```python
from sklearn.svm import SVC
from sklean
from sklearn.metrics import accuracy_score

clf = SVC(gamma='auto')
# make_pipeline(StandardScaler(), SVC(gamma='auto'))
clf.fit(X_train, y_train)
df_pred = clf.predict(X_test)

print(accuracy_score(y_test, df_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

dt_clf = DecisionTreeClassifier(random_state=151)
rf_clf = RandomForestClassifier(random_state=151)
lr_clf = LogisticRegression(solver='liblinear')

dt_clf.fit(X_train,y_train)
dt_pred = dt_clf.predict(X_test)
print('DecisionTree 정확도 : {0:.4f}'.format(accuracy_score(y_test,dt_pred)))

rf_clf.fit(X_train,y_train)
rf_pred = rf_clf.predict(X_test)
print('Randomforest 정확도 : {0:.4f}'.format(accuracy_score(y_test,rf_pred)))

lr_clf.fit(X_train,y_train)
lr_pred = lr_clf.predict(X_test)
print('LogisticRegression 정확도 : {0:.4f}'.format(accuracy_score(y_test,lr_pred)))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

clf = GradientBoostingClassifier(n_estimators=100, learning_rate=1.0, max_depth=1, random_state=0).fit(X_train, y_train)
clf.score(X_test, y_test)
```
